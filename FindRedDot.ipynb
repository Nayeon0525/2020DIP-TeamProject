```
boundaries = [
	([110, 25, 25], [200, 56, 50]),
]
```

```
for (lower, upper) in boundaries:
	# create NumPy arrays from the boundaries
	lower = np.array(lower, dtype = "uint8")
	upper = np.array(upper, dtype = "uint8")
	# find the colors within the specified boundaries and apply
	# the mask
	mask = cv2.inRange(test, lower, upper)
	output = cv2.bitwise_and(test, test, mask = mask)
	# show the images
	plt.figure(figsize=(19.20, 10.80))
	plt.imshow(output)
```

```
k = cv2.getStructuringElement(cv2.MORPH_RECT, (4,4))
opening = cv2.morphologyEx(output,cv2.MORPH_OPEN,k)
closing = cv2.morphologyEx(output,cv2.MORPH_CLOSE,k)
erosion = cv2.erode(output, k)
dilation = cv2.dilate(output, k)

plt.figure('ori vs. open vs. close', figsize=(12,6))
plt.subplot(1,3,1)
plt.axis('off')
plt.imshow(output,cmap='gray')

plt.subplot(1,3,2)
plt.axis('off')
plt.imshow(opening,cmap='gray')

plt.subplot(1,3,3)
plt.axis('off') 
plt.imshow(closing, cmap='gray')

plt.show()
```
